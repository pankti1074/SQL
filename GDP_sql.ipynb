```python
import pandas as pd
import sqlite3
```

```python
def rsql(query):
    return pd.read_sql_query(query, conn)
```

```python
#collect screener data as a data frame
imf_data = pd.read_html('https://www.imf.org/en/Publications/WEO/weo-database/2025/april/weo-report?c=513,514,516,522,924,819,534,536,826,544,548,556,867,868,948,518,836,558,565,853,566,862,813,524,578,537,866,869,846,582,&s=NGDPD,&sy=2020&ey=2025&ssm=0&scsm=1&scc=0&ssd=1&ssc=0&sic=0&sort=country&ds=.&br=1')
type(imf_data)
```

```python
# number of tables
len(imf_data)
```

```python
type(imf_data[0])
```

```python
imfdf = imf_data[0]
imfdf.head(3)
```

```python
conn = sqlite3.connect("imf.db")
c = conn.cursor()
```

```python
imfdf.columns
```

```python
c.execute(
    """ CREATE TABLE IF NOT EXISTS imfdata(
   'Country' text, 'Subject Descriptor' text, 'Units' text, 'Scale' text,
       'Country/Series-specific Notes' text, '2020' float, '2021' float, '2022' float, '2023' float, '2024' float,
       '2025' float);""")
```

```python
imfdf.to_sql("imfdata",conn, if_exists = "replace", index = False)
```

#### View the entire table


```python
rsql("SELECT * FROM imfdata")
```

#### What is the GDP of India?

```python
rsql("SELECT * FROM imfdata WHERE Country = 'India'")
```

#### Total of all GDP

```python
rsql("SELECT Country, SUM([2020] + [2021] + [2022] + [2023] + [2024] + [2025]) AS total_all_years FROM imfdata GROUP BY Country")
```

#### Average of last 5 years' GDP grouped by the country

```python
rsql(" SELECT Country, AVG(([2020] + [2021] + [2022] + [2023] + [2024] + [2025]) / 6.0) AS avg_all_years FROM imfdata GROUP BY Country")
```

#### Top 5 countries with the highest GDP in 2025

```python
rsql("SELECT Country, [2025] AS GDP_usd_billion FROM imfdata ORDER BY [2025] DESC LIMIT 5")
```

#### View the first 3 records

```python
rsql("SELECT * FROM imfdata LIMIT 3")
```

#### Rename columns

```python
rsql("SELECT Country AS Country_Name, [2025] AS GDP_in_2025 FROM imfdata LIMIT 5")
```

#### Top Countries with GDP Over $1000B in 2025

```python
rsql("SELECT Country, [2025] AS GDP FROM imfdata WHERE [2025] > 1000 ORDER BY [2025] DESC")
```

#### Countries with average GDP (2020 to 2025) > $250B

```python
rsql("SELECT Country, AVG([2020] + [2021] + [2022] + [2023] + [2024] + [2025]) / 6 AS avg_GDP FROM imfdata GROUP BY Country HAVING avg_gdp > 250 ORDER BY avg_GDP DESC")
```

#### Countries with 2025 GDP Between 1,000 and 10,000 Billion USD

```python
rsql("SELECT Country, [2025] AS GDP_2025 FROM imfdata WHERE [2025] BETWEEN 1000 AND 10000 ORDER BY [2025] DESC")
```

#### GDP Aggregates

```python
rsql("SELECT COUNT(*) AS total_entries, AVG([2025]) AS avg_gdp_2025, SUM([2025]) AS total_gdp_2025 FROM imfdata")
```

#### Unique units used in the dataset

```python
rsql("SELECT DISTINCT Units FROM imfdata")
```

#### Adjust a GDP value

```python
c.execute("UPDATE imfdata SET [2020] = 2675 WHERE Country = 'India'")
conn.commit()
```

```python
rsql("SELECT [2020] FROM imfdata WHERE Country = 'India'")
```

#### Inserting a tuple

```python
c.execute("""
INSERT INTO imfdata 
(Country, [Subject Descriptor], Units, Scale, [Country/Series-specific Notes], [2020], [2021], [2022], [2023], [2024], [2025]) 
VALUES 
('New_Country', 'Gross domestic product, current prices', 'U.S. dollars', 'Billions', NULL, 100.0, 120.5, 140.0, 160.0, 180.0, 200.0)
""")
conn.commit()
```

```python
rsql("SELECT * FROM imfdata WHERE Country = 'New_Country'")
```

#### All countries with names starting with "B"

```python
rsql("SELECT Country, [2025] FROM imfdata WHERE Country LIKE 'B%'")
```
